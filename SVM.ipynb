```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.metrics  import accuracy_score
```

```python
data=pd.read_csv('diabetes_dataset.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
mode_alcohol = data['Alcohol_Consumption'].mode()[0]
data['Alcohol_Consumption'] = data['Alcohol_Consumption'].apply(lambda x: mode_alcohol if pd.isnull(x) else x)
```

```python
data['diabetes']=(data['Fasting_Blood_Glucose']>=126).astype(int)
```

```python
numerical_cols=data.select_dtypes(include=['float64','int64']).columns
categorical_cols=data.select_dtypes(include=['object']).columns
```

```python
X=data.drop(columns=['diabetes'])
Y=data['diabetes']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
X_train.shape,Y_train.shape,X_test.shape,Y_test.shape
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numerical_cols),
    ('cat', OneHotEncoder(), categorical_cols)
])
```

```python
svm_pipeline = Pipeline(steps=[
    ('preprocessing', preprocessor),
    ('classifier', SVC(kernel='linear', C=1.0))
])
```

```python
svm_pipeline.fit(X_train,Y_train)
```

```python
accuracy = svm_pipeline.score(X_test, Y_test)
print(f'SVM Model Accuracy: {accuracy * 100:.2f}%')
```

```python
accuracy = svm_pipeline.score(X_train, Y_train)
print(f'SVM Model Accuracy: {accuracy * 100:.2f}%')
```
